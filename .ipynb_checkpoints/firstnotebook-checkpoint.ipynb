```python
work_hours = [('a',1000),('b',200),('c',800)]
```

```python
def find_best_emp(tup_list):
    max_hour=0
    best_emp=''
    for emp,hour in tup_list:
        if hour>max_hour:
            max_hour = hour
            best_emp = emp
        else:
            pass
    return f"Best employee is {best_emp} with {max_hour} working hours"

find_best_emp(work_hours)
            
```

```python
import random
num = random.randint(1,100)
guess_num = int(input("Enter a number"))
if guess_num == num:
    guess_counter.append(guess_num)
last_guess = 1000
guess_counter= []
while guess_num!=num:
    if guess_num > 100 or guess_num<1 :
        print ("Out of Bounds")
    elif len(guess_counter)==0 and abs(guess_num-num)<= 10 :
        guess_counter.append(guess_num)
        last_guess=guess_num
        print("WARM")
    elif len(guess_counter)==0 and abs(guess_num-num) >10 :
        guess_counter.append(guess_num)
        last_guess=guess_num
        print("COLD")
    elif  abs(guess_num-num) < abs(last_guess-num):
        guess_counter.append(guess_num)
        last_guess=guess_num
        print("WARMER")
    elif  abs(guess_num-num) >= abs(last_guess-num):
        guess_counter.append(guess_num)
        last_guess=guess_num
        print("COLDER")
    print(guess_counter)
    guess_num = int(input("Enter a number"))

print(guess_counter)
print(f"It took {len(guess_counter)} guesses")
        
```

```python
# Better code
import random

num = random.randint(1,100)

print("WELCOME TO GUESS ME!")
print("I'm thinking of a number between 1 and 100")
print("If your guess is more than 10 away from my number, I'll tell you you're COLD")
print("If your guess is within 10 of my number, I'll tell you you're WARM")
print("If your guess is farther than your most recent guess, I'll say you're getting COLDER")
print("If your guess is closer than your most recent guess, I'll say you're getting WARMER")
print("LET'S PLAY!")

guesses = [0]

    
while True:

    # we can copy the code from above to take an input
    guess = int(input("I'm thinking of a number between 1 and 100.\n  What is your guess? "))
    
    if guess < 1 or guess > 100:
        print('OUT OF BOUNDS! Please try again: ')
        continue
    
    # here we compare the player's guess to our number
    if guess == num:
        print(f'CONGRATULATIONS, YOU GUESSED IT IN ONLY {len(guesses)} GUESSES!!')
        break
        
    # if guess is incorrect, add guess to the list
    guesses.append(guess)
    
    # when testing the first guess, guesses[-2]==0, which evaluates to False
    # and brings us down to the second section
    
    if guesses[-2]:  
        if abs(num-guess) < abs(num-guesses[-2]):
            print('WARMER!')
        else:
            print('COLDER!')
   
    else:
        if abs(num-guess) <= 10:
            print('WARM!')
        else:
            print('COLD!')
```
